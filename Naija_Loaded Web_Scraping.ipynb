# Scraping this blog for it's latest news post, link, author and date.

```python
import pprint
import requests
from bs4 import BeautifulSoup

res = requests.get('https://www.naijaloaded.com.ng/news')
soup = BeautifulSoup(res.text, 'html.parser')

links = soup.select('h3')#[1].getText().replace('\n', '')
authors = soup.find_all("span", class_="author vcard" )#[0].getText()
dates = soup.select('time')#[0].get('datetime').replace('T', ", Time ")

def NaijaLoaded (links, Dates):
    NL = []
    for indx, item in enumerate(links):
        if indx <30:
            title = links[indx].getText().replace('\n', '')
            href = links[indx].select('a')[0].get('href')
            author = authors[indx].getText()
            Dates = dates[indx].get('datetime').replace('T', ", Time ")
            NL.append({'Title': title, 'Link': href, 'Dates': Dates, 'Author':author })
        else:
            return NL


NaijaLoaded (links, dates)

```
