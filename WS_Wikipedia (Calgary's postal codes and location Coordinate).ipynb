```python
# IMPORTING NECESSARY LIBRARIES

from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
import pandas as pd
```

```python
# DEFINING URL FOR SCRAPING CALGARY'S POSTAL-CODE, FROM WIKIPEDIA

url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_T"
```

The data from the HTML table was scraped into a DataFrame using (read_html)

```python
dataframe_list = pd.read_html(url, flavor='bs4')
```

Dataframe containing Postal Code, Borough, Neighborhood, latitude and Longitude

```python
dataframe_list[1]
```

```python
Alberta = dataframe_list[1] #Renaming the dataframe
```

```python
Alberta.info() # Getting the infomation of the dataframe
```

Retrieving our Area Of Interest (AOI) which is Calgary

```python
Calgary = Alberta[Alberta['Borough'].str.contains("Calgary")]
```

```python
# Resetting index after selecting our AOI from the ALberta dataframe

Calgary = Calgary.reset_index(drop=True)
Calgary
```

Calgary has 35 Neighborhoods

As seen SouthEast Calgary doesn't have assigned location, Geolocator will be used to obtain its location coordinates.

```python
#Importing library

from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
```

```python
# Obtaining lat and long data using geolocator

address = "Southeast Calgary"
geolocator = Nominatim(user_agent="foursquare_agent")
location = geolocator.geocode(address)
SE_latitude = location.latitude
SE_longitude = location.longitude
print (SE_latitude, SE_longitude)
```

```python
# Replacing 'Not assigned' with lat and long values

Calgary = Calgary.replace({'Latitude': 'Not assigned'}, str(SE_latitude))
Calgary = Calgary.replace({'Longitude': 'Not assigned'}, str(SE_longitude))
```

```python
Calgary['Latitude'] = Calgary['Latitude'].astype(float)
Calgary['Longitude'] = Calgary['Longitude'].astype(float)
```
