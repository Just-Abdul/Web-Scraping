## Web Scraping NetNaija website for their Latest Movies.
### It contains the Movie Title, Link and the date it was uploaded.

```python
import pprint
import requests
from bs4 import BeautifulSoup

res = requests.get('https://www.thenetnaija.com/videos/movies')

soup = BeautifulSoup(res.text, 'html.parser')
links = soup.select('.file-name')
date = soup.select('.other-info')

def Netnaija (links, date):
    
    Netnaija = []
    
    for indx, itm in enumerate(links):
        title = links[indx].getText()
        href = links[indx].find_all('a')[0].get('href')
        date = soup.select('.other-info')[indx].getText().replace('\n Mar ', '').replace('\n Feb ','')
        date = int(date)
        Netnaija.append({'Title':title, 'Link':href, 'Date':date})
    
    return Netnaija

Netnaija (links, date)
```

# Scraping the first 3 pages 

```python
import pprint
import requests
from bs4 import BeautifulSoup

res = requests.get('https://www.thenetnaija.com/videos/movies')
res2 = requests.get('https://www.thenetnaija.com/videos/movies/page/2')
res3 = requests.get('https://www.thenetnaija.com/videos/movies/page/3')


soup = BeautifulSoup(res.text, 'html.parser')
soup2 = BeautifulSoup(res2.text, 'html.parser')
soup3 = BeautifulSoup(res3.text, 'html.parser')


links = soup.select('.file-name')
links2 = soup2.select('.file-name')
links3 = soup3.select('.file-name')

multi_links = links + links2 + links3

def Netnaija (links):
    
    
    Netnaija = []
    
    
    for indx, itm in enumerate(links):
        title = links[indx].getText()
        href = links[indx].find_all('a')[0].get('href')
        
        Netnaija.append({'Title':title, 'Link':href})
    
    
    
    return Netnaija

Netnaija (multi_links)   
```
